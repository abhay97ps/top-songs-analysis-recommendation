```python
import time
def getLyrics(artist,song):
    try:
        gObj = genius.search_song(song, artist)
        if gObj is not None:
            L = gObj.lyrics
            return (1, L)
        return (0,"")
    except:
        time.sleep(60)
        return getLyrics(artist,song)
```

```python
import nltk
import nltk.sentiment.vader
from textstat.textstat import textstat as ts

# repeated lines
def countDup(lyrics):
    list = lyrics.split("\n")
    count = 0
    seen = set()
    for item in list:
        if item in seen:
            count += 1
        else:
            seen.add(item)
    return count

# sentiment
sent_analyzer = nltk.sentiment.vader.SentimentIntensityAnalyzer()

# combine with textstat features and return
def getLyricFeat(lyrics):
    l_mod = lyrics.replace("\n", ". ")
    st = sent_analyzer.polarity_scores(l_mod)
    # 'neg': 0.09, 'neu': 0.618, 'pos': 0.292, 'compound': 0.9874
    return {
        "sentiment_comp": st.get('compound'),
        "sentiment_pos": st.get('pos'),
        "sentiment_neu": st.get('neu'),
        "sentiment_neg": st.get('neg'),
        "f_k_grade" : ts.flesch_kincaid_grade(l_mod),
        "flesch_index" : ts.flesch_reading_ease(l_mod),
        "fog_index" : ts.gunning_fog(l_mod),
        "difficult_words" : ts.difficult_words(l_mod),
        "num_syllables" : ts.syllable_count(l_mod),
        "num_words" : ts.lexicon_count(l_mod, True),
        "num_lines" : ts.sentence_count(l_mod),
        "num_dup" : countDup(lyrics)
    }
```

```python
import re
import os

def cleanLyrics(lyrics):
    # remove [Chorus] [Verse 1] tags
    lyrics = re.sub('\[.*?\]', '', lyrics)
    # remove repeaters eg (2x)
    lyrics = re.sub('\(.*?\)', '', lyrics)
    # remove empty line and trailing Whitespaces
    lyrics = os.linesep.join([s for s in lyrics.splitlines() if s.strip()])
    return lyrics
```

```python
def writeLyrics(path, text):
    with open(path, "w") as textFile:
        textFile.write(text)
```
