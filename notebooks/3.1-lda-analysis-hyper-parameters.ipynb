```python
import pandas as pd
```

```python
path_lda_anal = "../data/final/lda_tuning_results.csv"
data = pd.read_csv(path_lda_anal)
```

```python
data.head()
data.columns = data.columns.str.strip()
```

```python
vs1 = data[data["Validation_Set"]=="75% Corpus"]
vs2 = data[data["Validation_Set"]=="100% Corpus"]
```

```python
# taking alpha and beta as constant at 0.01 and 0.01 respectively
opt_topics_set_1 = vs1[(vs1["Alpha"]=="0.01") & (vs1["Beta"]=="0.31")]
opt_topics_set_2 = vs2[(vs2["Alpha"]=="0.01") & (vs2["Beta"]=="0.31")]
```

```python
opt_topics_set_1.head()
```

```python
opt_topics_set_1.plot.line(x="Topics", y="Coherence")
```

```python
opt_topics_set_2.plot.line(x="Topics", y="Coherence")
```

```python
# need to visibility as chaning beta and alpha drastically change results
topic_group_mean = vs1.groupby(["Topics"]).mean().reset_index()
topic_group_mean.plot.line(x="Topics", y="Coherence")
```

```python
topic_group_mean = vs2.groupby(["Topics"]).mean().reset_index()
topic_group_mean.plot.line(x="Topics", y="Coherence")
```

```python
topic_group_mean = vs1.groupby(["Alpha"]).mean().reset_index()
topic_group_mean.plot.line(x="Alpha", y="Coherence")
```

```python
topic_group_mean = vs2.groupby(["Alpha"]).mean().reset_index()
topic_group_mean.plot.line(x="Alpha", y="Coherence")
```

```python
topic_group_mean = vs1.groupby(["Beta"]).mean().reset_index()
topic_group_mean.plot.line(x="Beta", y="Coherence")
```

```python
topic_group_mean = vs2.groupby(["Beta"]).mean().reset_index()
topic_group_mean.plot.line(x="Beta", y="Coherence")
```
